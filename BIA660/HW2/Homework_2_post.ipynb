- Q1. Write a function to scrape all **reviews on the first page**, including, 
    - **reviewer's name** (see (1) in Figure)
    - **rating** (see (2) in Figure)
    - **date** (see (3) in Figure)
    - **review content** (see (4) in Figure. For each review text, need to get the **complete text**, i.e., need to expand the `more` button. Only text is needed, pictures are not needed. Hint: take a close look at the content of the html file.)
    - **likes** (see (5) in Figure). 
    - If a field, e.g. score, is missing, use `None` to indicate it. 
- `Function Input`: book page URL
- `Function Output`: save all reviews as a DataFrame of columns (`reviewer, rating, date, review, like`). E.g., for the given URL, you can get 30 reviews.(10 points)
- **Note**: GoodReads occasionaly blocks request. You may get an error that is not due to your codes. You may try a couple of times.

```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
 

def getReviews(page_url):

   # add your code here
    review = pd.DataFrame()

    page = requests.get(page_url)
    
    if page.status_code==200:      

        soup = BeautifulSoup(page.content, 'html.parser')

        # locate user review area
        for soup_new in soup.find_all('div', {"class":"friendReviews elementListBrown"}):

            # create temp dict to contain all info
            user_info = {}

            # 1.reviewer's name
            user_info['name'] = soup_new.find('a', {"class":"user"}).text

            # 2.rating
            # handle the maybe lack of rating
            if soup_new.find('span', {"class":"staticStars notranslate"}) == None:
                user_info['rating'] = 'None'
            else:
                user_info['rating'] = soup_new.find('span', {"class":"staticStars notranslate"}).text

            # 3.date
            user_info['date'] = soup_new.find('a', {"class":"reviewDate createdAt right"}).text

            # 4.review content
            # no need to click 'more' button using BeautifulSoup 
            # remove line breaks '\n', expand button text '...more' at the end
            user_info['review']= soup_new.find('div', {"class":"reviewText stacked"}).text.replace('\n','').replace('\r','').replace('...more','')

            # 5.likes
            # remove 'likes-', No need to leave that dot at the end, like counts are more useful as int I suppose
            like = soup_new.find('span', {"class":"likesCount"}).text
            tmp = ''.join(filter(str.isdigit, like))
            user_info['like'] = int(tmp)

            # update dataframe
            review = review.append(user_info, ignore_index=True)
        
    return review  
    
```

```python
# I choose <Wonder> by R.J. Palacio
page_url = 'https://www.goodreads.com/book/show/11387515-wonder?from_search=true&from_srp=true&qid=QRjldjMRu4&rank=1'
reviews = getReviews(page_url)
reviews
```
